#Pregunta 1. ¬øQu√© categor√≠as de alimentos contribuyen m√°s a la ingesta cal√≥rica total? ìåâ‚óØìáã
*Hip√≥tesis:* Las categor√≠as de alimentos como 'Carnes' y 'Grasas s√≥lidas' contribuyen m√°s a la ingesta cal√≥rica total en comparaci√≥n con las categor√≠as de 'Vegetales' y 'Frutas'.


> *Para realizar √©sta hip√≥tesis creamos un grafico de barras que nos muestre un promedio de total de calor√≠as por cada categor√≠a de alimentos. Evitamos el uso de valores totales para que el gr√°fico sea mas acertivo*


---




```python
# Creamos una lista de las categor√≠as que necesitamos
categories = ['Vegetales', 'Frutas', 'Carnes', 'Grasas']

# Renombramos las columnas en el DataFrame
df_food.rename(columns={'Vegetables': 'Vegetales', 'Fruits': 'Frutas', 'Meats': 'Carnes', 'Solid_Fats': 'Grasas'}, inplace=True)

# Creamos una nueva columna 'Category' basada en las columnas existentes
df_food['Category'] = df_food[categories].idxmax(axis=1)

# Filtramos solo las categor√≠as deseadas
filtered_df = df_food[df_food['Category'].isin(categories)]

# Agrupamos por categor√≠a y calculamos el promedio de calor√≠as
average_calories_by_category = filtered_df.groupby('Category')['Calories'].mean().reset_index()

# Ordenamos el DataFrame de mayor a menor seg√∫n las calor√≠as
average_calories_by_category = average_calories_by_category.sort_values('Calories', ascending=False)

# Creamos el gr√°fico de barras
fig = px.bar(average_calories_by_category,
             x='Category',
             y='Calories',
             title='Promedio de Ingesta Cal√≥rica por Categor√≠a de Alimentos',
             labels={'Calories':'Promedio de Ingesta Cal√≥rica', 'Category':'Categor√≠a de Alimentos'},
             color='Calories',
             color_continuous_scale='Portland')

# Actualizamos el dise√±o para mantener el orden de las categor√≠as y ajustar la barra de color
fig.update_layout(
    xaxis={'categoryorder':'total descending'},
    coloraxis_colorbar=dict(title='Calor√≠as')
)

# Mostramos el gr√°fico
fig.show()
```



>Con base en los resultados podemos concluir que las categor√≠as de alimentos como 'Carnes' y 'Grasas' contribuyen de manera m√°s significativa a la ingesta cal√≥rica total en comparaci√≥n con las categor√≠as de 'Vegetales' y 'Frutas'. Estos datos confirman la hip√≥tesis de que las categor√≠as m√°s altas en calor√≠as est√°n asociadas con alimentos de origen animal y alto contenido graso. ‚úÖ



#Pregunta 2: ¬øLos alimentos que contienen soja tienden a tener menos grasas saturadas en comparaci√≥n con alimentos que contienen carne? ü•©üå±

Hip√≥tesis: Los alimentos que contienen soja tienden a tener un menor contenido de grasas saturadas en comparaci√≥n con los alimentos que contienen carne.

> Para validar esta hip√≥tesis, creamos un gr√°fico de torta que compara la cantidad total de grasas saturadas en alimentos con soja y carne, mostrando de manera clara las diferencias entre ambas categor√≠as.



```python
# Filtramos los datos
soy_data = df_food[df_food['Soy'] > 0]
meats_data = df_food[df_food['Meats'] > 0]

# Creamos un nuevo DataFrame para el gr√°fico de torta
pie_data = pd.DataFrame({
    'Category': ['Soja', 'Carnes'],
    'Saturated_Fats': [soy_data['Saturated_Fats'].sum(), meats_data['Saturated_Fats'].sum()]
    })

# Creamos el gr√°fico de torta
fig = px.pie(pie_data,
             values='Saturated_Fats',
             names='Category',
             title='Comparaci√≥n de Grasas Saturadas entre Alimentos con Soja y Carnes',
             color_discrete_sequence=['rgb(255, 0, 0)', 'rgb(0, 128, 0)'])

# Actualizamos el dise√±o para mejorar la presentaci√≥n
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(
    legend_title="Categor√≠a",
    font=dict(size=14),
    title=dict(font=dict(size=18))
)

# Mostramos el gr√°fico
fig.show()
```

> Con base en los resultados obtenidos podemos concluir que los alimentos que contienen soja tienden a tener un contenido significativamente menor de grasas saturadas en comparaci√≥n con los alimentos que contienen carne. Esto respalda la hip√≥tesis planteada, sugiriendo que la soja es una opci√≥n m√°s baja en grasas saturadas frente a la carne. ‚úÖ








# Pregunta 3: ¬øC√≥mo influyen los az√∫cares a√±adidos en la ingesta cal√≥rica de diferentes categor√≠as de alimentos?  ü§îüîé

---

Hip√≥tesis: Los az√∫cares a√±adidos tienen una influencia significativa en el aumento de la ingesta cal√≥rica en las categor√≠as de alimentos que los contienen, en comparaci√≥n con las categor√≠as de alimentos sin az√∫cares a√±adidos.

>Para validar esta hip√≥tesis, generamos un gr√°fico de barras que compara la ingesta cal√≥rica promedio de diferentes categor√≠as de alimentos, resaltando la influencia de los az√∫cares a√±adidos en cada una de ellas.

```python
# Definimos las categor√≠as de alimentos que queremos analizar
food_categories = ['Grains', 'Whole_Grains', 'Vegetables', 'Fruits', 'Milk', 'Meats', 'Soy']

# Creamos un nuevo DataFrame con las columnas que nos interesan
df_analysis = df_food[food_categories + ['Added_Sugars', 'Calories']]

# Transformamos el DataFrame para tener una fila por cada categor√≠a de alimento
df_melted = df_analysis.melt(id_vars=['Added_Sugars', 'Calories'],
                             var_name='Food_Category',
                             value_name='Amount')

# Filtramos para mantener solo las filas donde Amount > 0
df_melted = df_melted[df_melted['Amount'] > 0]

# Agrupamos por categor√≠a de alimento y calculamos el promedio de az√∫cares a√±adidos y calor√≠as
df_grouped = df_melted.groupby('Food_Category').agg({
    'Added_Sugars': 'mean',
    'Calories': 'mean'
}).reset_index()

# Creamos un diccionario para mapear los nombres de las categor√≠as
category_map = {
    'Milk': 'Leche', 'Meats': 'Carnes', 'Grains': 'Granos', 'Fruits': 'Frutas', 'Soy': 'Soja', 'Vegetables': 'Vegetales', 'Whole_Grains': 'Granos Integrales'}

# Aplicamos el mapeo a la columna 'Food_Category'
df_grouped['Food_Category'] = df_grouped['Food_Category'].map(category_map)

# Ordenamos el DataFrame de mayor a menor seg√∫n las calor√≠as
df_grouped = df_grouped.sort_values('Calories', ascending=False)

# Creamos el gr√°fico de barras
fig = px.bar(df_grouped,
             x='Food_Category',
             y='Calories',
             color='Added_Sugars',
             color_continuous_scale=px.colors.sequential.Rainbow,
             labels={'Food_Category': 'Categor√≠a de Alimento',
                     'Calories': 'Calor√≠as Promedio',
                     'Added_Sugars': 'Az√∫cares A√±adidos Promedio'},
             title='Influencia de los Az√∫cares A√±adidos en la Ingesta Cal√≥rica por Categor√≠a de Alimento')

# Personalizamos el dise√±o
fig.update_layout(
    xaxis_title='Categor√≠a de Alimento',
    yaxis_title='Calor√≠as Promedio',
    coloraxis_colorbar_title='Az√∫cares A√±adidos Promedio'
)

# Actualizamos el orden de las categor√≠as en el eje x
fig.update_xaxes(categoryorder='total descending')

# Mostramos el gr√°fico
fig.show()
```

>Los resultados muestran que la relaci√≥n entre los az√∫cares a√±adidos y la ingesta cal√≥rica var√≠a seg√∫n la categor√≠a de alimentos. Aunque algunos alimentos con mayor cantidad de az√∫cares a√±adidos presentan una ingesta cal√≥rica significativa, no existe una correlaci√≥n directa y consistente entre ambos factores en todas las categor√≠as. Esto sugiere que, aunque los az√∫cares a√±adidos influyen en la ingesta cal√≥rica, otros componentes nutricionales tambi√©n juegan un papel importante en la determinaci√≥n del contenido cal√≥rico de los alimentos. En conclusi√≥n, podemos decir que nuestra hip√≥tesis no es v√°lida. ‚ùå
